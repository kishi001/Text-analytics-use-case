```python
import requests
page = requests.get("https://weibo.com/login.php?ssl_rnd=1510308132.3388")
contents = page.content

from bs4 import BeautifulSoup
soup = BeautifulSoup(contents, 'html.parser')

for i in soup.find_all('a'):
    print(i.text)
```
